```python
import numpy as np
data = np.array([
    [100, 30],
    [ 80, 20],
    [ 90, 26],
    [ 50, 15],
    [ 70, 20],
    [ 80, 22],
    [ 75, 23],
    [100, 15],
    [100, 35],
    [ 20, 21]
])
x = data[:, 1].reshape(-1, 1)  
y = data[:, 0]                
```

```python
print(data)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.scatter(x, y)
plt.xlabel("study")
plt.ylabel("score")
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
print(regr.coef_)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = regr.predict(x)
r2_score(y, y_pred)
```

```python
plt.scatter(x, y, color='pink')
plt.plot(x, y_pred, color='purple', linewidh=3)
plt.xlabel("study")
plt.ylabel("score")
plt.show()
```
