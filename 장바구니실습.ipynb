```python
#연관규칙분석 실습
```

```python
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [
    ['Milk', 'Cookie', 'Apple', 'Beans', 'Eggs', 'Yogurt'],
    ['Coke', 'Cookie', 'Apple', 'Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Orange', 'Corn', 'Beans', 'Yogurt'],
    ['Corn', 'Cookie', 'Cookie', 'Beans', 'Ice cream', 'Eggs']
]
```

```python
type(dataset) 
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
```

```python
df
```

```python
#mlxtend의 Apriori 알고리즘 
from mlxtend.frequent_patterns import association_rules
apriori(df, min_support = 0.6)
```

```python
apriori(df, min_support = 0.6, use_colnames = True)
```

```python
#결과 필터링해보기
```

```python
frequent_itemsets = apriori = apriori(df, min_support = 0.6, use_colnames = True)
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 2) & (frequent_itemsets['support']>= 0.6)]
```

```python
#연관규칙분석결과에서 confidence, lift 등 조건에 맞는 패턴 찾기 
```

```python
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
rules
```

```python
rules2 = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1.2)
rules2
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
```

```python
rules[(rules['antecedent_len'] >= 2) & (rules['confidence'] > 0.75) & (rules['lift'] > 1.2)]
```
